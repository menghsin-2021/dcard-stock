```python
# import package
from datetime import date,timedelta
import plotly.graph_objects as go
import plotly
import pandas as pd
import numpy as np
```

```python
df_total_like_count = pd.read_csv('dcard_stock/data/df_posts_total_like_count.csv', index_col=0, parse_dates=True)
df_total_like_count_shipping = pd.read_csv('dcard_stock/data/df_posts_total_like_count_shipping.csv', index_col=0, parse_dates=True)
df_2603 = pd.read_csv('dcard_stock/data/df_2603_price_20201121.csv', index_col=0, parse_dates=True)
df_2609 = pd.read_csv('dcard_stock/data/df_2609_price_20201121.csv', index_col=0, parse_dates=True)
df_2615 = pd.read_csv('dcard_stock/data/df_2615_price_20201121.csv', index_col=0, parse_dates=True)
```

```python
# https://plotly.com/python/line-and-scatter/

fig = go.Figure()

fig.add_trace(go.Scatter(x=df_total_like_count.index,
           y=df_total_like_count['total_like_count'],
           mode='lines',
           opacity=0.7,
           name='total_like_count',
           textposition='bottom center'))

fig.add_trace(go.Scatter(x=df_total_like_count_shipping.index,
           y=df_total_like_count_shipping['total_like_count'],
           mode='lines',
           opacity=0.7,
           name='shipping_total_like_count',
           textposition='bottom center'))

fig.show()
```

```python
# https://plotly.com/python/line-and-scatter/

fig = go.Figure()

fig.add_trace(go.Scatter(x=df_2603.index,
           y=df_2603['收盤價'],
           mode='lines',
           opacity=0.7,
           name='收盤價',
           textposition='bottom center'))

fig.add_trace(go.Scatter(x=df_2609.index,
           y=df_2609['收盤價'],
           mode='lines',
           opacity=0.7,
           name='收盤價',
           textposition='bottom center'))

fig.add_trace(go.Scatter(x=df_2615.index,
           y=df_2615['收盤價'],
           mode='lines',
           opacity=0.7,
           name='收盤價',
           textposition='bottom center'))


fig.show()
```

```python
from plotly.subplots import make_subplots

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])


# Add traces
fig.add_trace(
    go.Scatter(x=df_total_like_count_shipping.index,
           y=df_total_like_count_shipping['total_like_count'],
           mode='lines',
           opacity=0.7,
           name='shipping_total_like_count',
           textposition='bottom center'),    
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=df_2603.index,
           y=df_2603['收盤價'],
           mode='lines',
           opacity=0.7,
           name='當日收盤價',
           textposition='bottom center'),
    secondary_y=True,
)

fig.add_trace(go.Scatter(x=df_2609.index,
           y=df_2609['收盤價'],
           mode='lines',
           opacity=0.7,
           name='收盤價',
           textposition='bottom center'),
    secondary_y=True)

fig.add_trace(go.Scatter(x=df_2615.index,
           y=df_2615['收盤價'],
           mode='lines',
           opacity=0.7,
           name='收盤價',
           textposition='bottom center'),
    secondary_y=True)

# Add figure title
fig.update_layout(
    title_text="Double Y Axis Example"    
)

# Set x-axis title
fig.update_xaxes(title_text="xaxis title")

# Set y-axes titles
fig.update_yaxes(title_text="<b>總按讚數(個)</b>", secondary_y=False)
fig.update_yaxes(title_text="<b>當日收盤價(元)</b>", secondary_y=True)


fig.show()
```
