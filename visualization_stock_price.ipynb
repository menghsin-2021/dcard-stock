```python
df2 = pd.read_csv('stock_price_raw/2609_20200101.csv')
df2['日期'] = [str(int(time.split('/')[0]) + 1911) + '-' + time.split('/')[1] + '-' + time.split('/')[2] for time in df2['日期']]
pd.to_datetime(df2['日期'], format='%Y-%m-%d')
df2.index=df2['日期']
df_2609_price = df2.loc['2020-11-21':'2021-08-19', ['收盤價']]
df_2609_price.to_csv('df_2609_price_20201121.csv', encoding='utf_8_sig')


```

```python
df3 = pd.read_csv('stock_price_raw/2615_20200101.csv')
df3['日期'] = [str(int(time.split('/')[0]) + 1911) + '-' + time.split('/')[1] + '-' + time.split('/')[2] for time in df3['日期']]
pd.to_datetime(df3['日期'], format='%Y-%m-%d')
df3.index=df3['日期']
df_2615_price = df3.loc['2020-11-21':'2021-08-19', ['收盤價']]
df_2615_price.to_csv('df_2615_price_20201121.csv', encoding='utf_8_sig')
```

```python
# https://plotly.com/python/line-and-scatter/

fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=df_2603_price.index,
           y=df_2603_price['收盤價'],
           mode='lines',
           opacity=0.7,
           name='closing_price_2603',
           textposition='bottom center'))
fig.add_trace(go.Scatter(x=df_2609_price.index,
           y=df_2609_price['收盤價'],
           mode='lines',
           opacity=0.7,
           name='closing_price_2609',
           textposition='bottom center'))
fig.add_trace(go.Scatter(x=df_2615_price.index,
           y=df_2615_price['收盤價'],
           mode='lines',
           opacity=0.7,
           name='closing_price_2615',
           textposition='bottom center'))


fig.show()
```
